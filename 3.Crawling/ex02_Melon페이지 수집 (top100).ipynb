- 데이터 내보내기
    1. 데이터 리스트에 담아주기
    2. 표와같은 형태로 만들어주기(DataFrame화)
    3. csv파일로 내보내기

```python
# 데이터 리스트에 담기
# 빈리스트 생성
song_list = []
singer_list = []
rank_list = []
# 반복문을 통해 데이터 누적
for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
song_list
```

```python
import pandas as pd
```

```python
# DataFrame화(dictionary 활용)
# 데이터프레임 만들 때 키값이 컬럼명으로 됨
dic = {'노래제목':song_list,'가수':singer_list,'순위':rank_list}
melon=pd.DataFrame(dic)
melon.set_index('순위', inplace=True)
melon
```

```python
# csv 파일로 내보내기
# df명.to_csv('생성할 경로/파일명.csv')
melon.to_csv('./data/멜론차트.csv',)
```
