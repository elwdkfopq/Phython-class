### 반복문(while문, for문)
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문

#### while문
- 반복 횟수가 명확하지 않을 때 활용
```python
while 조건식:
    코드블럭
    코드블럭
```

````python
# while문을 활용해서 "파이썬 최고!!"를 13번 출력
cnt=1
while cnt<=13:
    print("파이썬 최고!!")
    cnt+=1
````

````python
# break - 반복문을 나가는 기능! if랑 세트로 많이 사용된다

while True:
    print("무한루프")
    break
````

````python
cnt=1
while True:
    print("파이썬 최고!!")
    if cnt==13:
        break
    cnt+=1
````

````python
# 두개의 정수를 입력받아서 더하는 코드를 작성
# 단 0과 0을 입력했을 떄는 프로그램 종료 출력
````

````python
while True:
    num1=int(input("첫번째 정수 입력>>"))
    num2=int(input("두번째 정수 입력>>"))
    if num1==0 and num2==0:
        print("프로그램이 종료되었습니다")
        break
    print(f"두 정수의 합>>{num1+num2}")
    
````

````python
weight=int(input("현재 몸무게:"))
goal=int(input("목표 몸무게:"))
cnt=0
while True:
    cnt+=1
    gamrang=int(input(f"{cnt}주차 감량 몸무게"))
    weight-=gamrang
    if weight<=goal:
        print(f"{weight}달성!! 축하합니다!")
        break
    
````

#### 2. for문
- 반복횟수가 명확할 때 활용
    1. 인덱스를 지정해서 반복하는 방법
    2. foreach 스타일로 리스트, 튜플 등 아이템의 갯수에 따라 반복하는 방법

````python
## range 함수 사용
## range(시작값, 끝값, 증가값)
for i in range(1,10,1):
    print(i, end=" ")# 줄바꿈 제거
````

````python
for i in range(10,1,-1):
    print(i, end=" ")
````

````python
## 2. foreach 스타일 반복문
## 주로 아이템이 들어있는 리스트, 튜플, 딕셔너리 등을 반복하고 싶을 때 사용
world_list = ["apple","banana","orange"]
for w in world_list: # 순차적으로 리스트의 아이템을 꺼내서 변수에 담음
    print(w, end=" ")
````

````python
# 문자열 타입은 각 글자마다 인덱스가 부여되기 때문에 반복문에서 활용 가능
for alphabet in "apple":
    print(alphabet, end=" ")
````

````python
## 아이템이 2개씩 묶인 튜플 반복하기
student_score = (('hhd',90),('pbk',88),('jh',50))
for student in student_score :
    print(student,end=" ")
````

````python
for name,score in student_score :
    print("name: ", name)
    print("score:", score)
````

````python
# foreach 스타일에서 인덱스를 같이 쓰고 싶은 경우
item_list = ['A','B','C','D','E']
for index, item in enumerate(item_list): #enumerate -> item 과 index를 같이 꺼내주는 함수 
    print(index, "번째", item)
````

````python
score_list=[90,45,70,60,55]
for index,score in enumerate(score_list):
    if score>=60:
        print(f"{index+1}번 학생은 합격입니다")
    elif score<60:
        print(f"{index+1}번 학생은 불합격입니다")
````

````python
sum=0
for num in range(1,101,1):
    if num%3==0:
        sum+=num
print(sum)
````

````python
# range에서 3의배수만 나오도록하기
sum_3 = 0
for num in range(3,100,3):
    sum_3 += num
print(sum_3)
````

````python
list2=[4,5,2,1,99,15,2,7,27]
smallest=100
for num in list2:
    if num<smallest:
        smallest=num
print(smallest)
````

````python
#데이터 로딩
import pickle

with open("./data/movie_review.pkl","rb") as f:
    movie_data = pickle.load(f)
````

````python
# 2. 5000개의 리뷰들의 평균 길이

summ=0
for review,_ in movie_data:
    summ+= len(review)
print(f"평균:{summ/len(movie_data)}")
````
