```python
# 갖고있는 돈을 검사해서 
# 10000원 이상인 경우 택시를 타고, 그렇지 않은 경우 걸어간다!!
# 10000원 미만인데 5000원 이상인 경우 버스를 타고 간다!!

money= int(input("갖고있는 돈을 입력하세요>>"))
if money >= 10000:
    print("택시를 탄다")
elif money >=5000:
    print("버스를 탄다!")
else :
    print("걸어간다!")
```

```python
num1 = int(input("첫 번째 정수 입력>>"))
num2 = int(input("두 번째 정수 입력>>"))
if num1>num2:
    print("첫 번쨰 정수가 더 큽니다.")
elif num1<num2:
    print("두 번째 정수가 더 큽니다.")
elif num1==num2:
    print("두 수가 똑같습니다.")
```

```python
score = int(input("점수입력>>"))
if score>100:
    print("잘못입력하셨습니다")
elif score>=90:
    print(f"{score}은 A학점 입니다.")
elif score>=80:
    print(f"{score}은 B학점 입니다.")
elif score>=70:
    print(f"{score}은 C학점 입니다.")
elif score>=60:
    print(f"{score}은 D학점 입니다.")
else:
    print(f"{score}은 F학점 입니다.")
```
